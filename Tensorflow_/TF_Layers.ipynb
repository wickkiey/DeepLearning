```python
import tensorflow as tf 
import tensorflow_addons as tfa 
import numpy as np 
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
# Read Image 

img = tf.io.decode_jpeg(tf.io.read_file('D:\digital_enhancement\images\kk\image_170.jpg'))
img = tf.image.resize(img,[72*2,128*2])
img = img[tf.newaxis,...]
img = tf.cast(img, tf.float32) / 255.
img.shape
```

```python
# Conv 2d
print("Input Image Shape: ", img.shape)
ip_img = img
conv2d = tf.keras.layers.Conv2D(3, 3,input_shape=(None, None, 3))
out = conv2d(ip_img)

# for i in range(2):
#     out = conv2d(out)

print("Output Image Shape: ", out.shape)
# plt.imshow(out.numpy())
```
